```python
# Read the Excel file into a pandas DataFrame
df_metadonnee_source = pd.read_excel(excel_file, sheet_name="Mapping Métadonnée-Source")

# Display the DataFrame
df_metadonnee_source
# Create a dictionary with key = Métadonnée and value = Sources
dict_metadonnee_source = dict(zip(df_metadonnee_source['Métadonnée'], df_metadonnee_source['Sources'].str.split(', ')))

# Display the dictionary
dict_metadonnee_source
```

```python
df_taches_bibliothecaire = pd.read_excel(excel_file, sheet_name="Scenario + Tache")
```

```python
df_filled = df_taches_bibliothecaire.fillna(method='ffill')
df_filled
```

```python
from collections import defaultdict

dict_tache_bibliotecaire = defaultdict(set)

for index, row in df_filled.iterrows():
    tache = row['Tache']
    bibliotecaire = str(row['Bibliothécaire'])
    if "Lyne Rajotte" in bibliotecaire: # elimination dupliqué de Lyne Rajotte qui a eu 2 entretiens
            bibliotecaire = "Lyne Rajotte"

    metadonnee_substring= " (Métadonnée)"
    if metadonnee_substring in tache:
        tache_trimmed = tache.replace(metadonnee_substring, "")
        source = dict_metadonnee_source[tache_trimmed]
        print(tache_trimmed, source)
        for s in source:
            dict_tache_bibliotecaire[s + " (Source)"].add(bibliotecaire)
    
    dict_tache_bibliotecaire[tache].add(bibliotecaire)
         
```

```python
dict_tache_bibliotecaire
```

```python
for key, value in sorted(dict_tache_bibliotecaire.items(), key=lambda x: len(x[1]), reverse=True):
    print(f"Key: {key:<60} Length of Values: {len(value):<5}")
```

```python
list(dict_tache_bibliotecaire.keys())
```

```python
markdown_table = "| Tache | Type | Nombre bibliothécaires |\n"
for key, value in sorted(dict_tache_bibliotecaire.items(), key=lambda x: len(x[1]), reverse=True):
    # print(f"Key: {key:<60} Length of Values: {len(value):<5}")
    if "(" in key:
        name, type = key.split(" (")
        type = type.replace(")", "")
        markdown_table += f"| {name} | {type} | {len(value)} | \n"
    else:
        markdown_table += f"| {key} | | {len(value)} | \n"
print(markdown_table)
# convertir le resultat en excel avec https://tableconvert.com/markdown-to-excel
# copier dans excel des recits utilisateurs, feuille "Nb bibliothécaires par tache"
```
