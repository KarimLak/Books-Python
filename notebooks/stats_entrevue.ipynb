```python
import pandas as pd

# Specify the path to the Excel file
excel_file = r"C:\Users\dre\OneDrive - Pole sur les données massives en culture\240816a - recits utilisateurs, comparaison avec outil existant.xlsx"

# Read the Excel file into a pandas DataFrame
df = pd.read_excel(excel_file)

# Display the DataFrame
df
```

```python
df_filled = df.fillna(method='ffill')
df_filled
```

```python
from collections import defaultdict

dict_tache_bibliotecaire = defaultdict(set)

for index, row in df_filled.iterrows():
    tache = row['Tache']
    source = row['Source'] if type(row['Source']) == str else "N/A"
    print(source)
    bibliotecaire = str(row['Bibliotécaire'])

    if "Lyne Rajotte" in bibliotecaire: # elimination dupliqué de Lyne Rajotte qui a eu 2 entretiens
        bibliotecaire = "Lyne Rajotte"
    dict_tache_bibliotecaire[tache].add(bibliotecaire)
    
    for source_temp in source.split(", "):
        print(source_temp)
        if source_temp != "N/A":
            dict_tache_bibliotecaire[source_temp].add(bibliotecaire)
```

```python
for key, value in sorted(dict_tache_bibliotecaire.items(), key=lambda x: len(x[1]), reverse=True):
    print(f"Key: {key:<60} Length of Values: {len(value):<5}")
```
