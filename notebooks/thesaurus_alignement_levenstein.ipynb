# basic theme matching
match if source theme name contained in thema theme name

```python
theme_evaluation_naive = {}
match_count = 0
for theme_to_evaluate in themes_validation_set_ground_truth.keys():
    for thema_theme in thema_code_dict.values():
        if theme_to_evaluate == thema_theme["CodeDescription"].lower() and theme_to_evaluate not in theme_evaluation_naive.values():
            print(theme_to_evaluate, "->", thema_theme)
            match_count += 1
            theme_evaluation_naive[theme_to_evaluate] = thema_theme["CodeValue"]
theme_evaluation_naive
```

```python
def calculate_metrics(theme_evaluation):
    correct_matches = 0

    for key, value in theme_evaluation.items():
        if key in themes_validation_set_ground_truth and value in themes_validation_set_ground_truth[key]:
            print(key, value, themes_validation_set_ground_truth[key])
            correct_matches += 1

    TP = correct_matches
    FP = len(theme_evaluation) - correct_matches
    FN = len(themes_validation_set_ground_truth) - correct_matches  # no result counts as a non-match

    accuracy = (TP / len(themes_validation_set_ground_truth)) * 100
    precision = TP / (TP + FP)
    recall = TP / (TP + FN)
    f1_score = 2 * (precision * recall) / (precision + recall)

    return precision, recall, f1_score, accuracy

precision, recall, f1_score, accuracy = calculate_metrics(theme_evaluation_naive)
precision, recall, f1_score, accuracy
```

```python
from fuzzywuzzy import fuzz

def calculate_similarity(sentences1, sentences2):
    similarity_pairs = []
    for sentence1 in sentences1:
        max_similarity = 0
        most_similar_sentence = ""
        for sentence2 in sentences2:
            similarity = fuzz.token_set_ratio(sentence1, sentence2)
            if similarity > max_similarity:
                max_similarity = similarity
                most_similar_sentence = sentence2
        similarity_pairs.append((sentence1, most_similar_sentence))
    return similarity_pairs

# Example usage
sentences1 = ["I love coding", "Python is awesome", "Machine learning is fascinating"]
sentences2 = ["I love programming", "Python is great", "Deep learning is interesting"]
similarity_pairs = calculate_similarity(sentences1, sentences2)
print(similarity_pairs)
```
