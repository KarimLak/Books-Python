```python
from fuzzywuzzy import fuzz

def calculate_similarity(sentences1, sentences2):
    similarity_pairs = []
    for sentence1 in sentences1:
        max_similarity = 0
        most_similar_sentence = ""
        for sentence2 in sentences2:
            similarity = fuzz.token_set_ratio(sentence1, sentence2)
            if similarity > max_similarity:
                max_similarity = similarity
                most_similar_sentence = sentence2
        similarity_pairs.append((sentence1, most_similar_sentence))
    return similarity_pairs

# Example usage
sentences1 = ["I love coding", "Python is awesome", "Machine learning is fascinating"]
sentences2 = ["I love programming", "Python is great", "Deep learning is interesting"]
similarity_pairs = calculate_similarity(sentences1, sentences2)
print(similarity_pairs)
```
