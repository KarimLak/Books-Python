# reviewers assignement
2 reviewers per sample for 3 reviwers

```python
def assign_reviewers(samples):
    reviewers = ['Juliette', 'Michel', 'Karim']

    # Create a cycle iterator for the reviewers
    reviewers_cycle = itertools.cycle(reviewers)

    # Assign reviewers to each sample with a UUID
    assignments = {}
    for sample in samples:
        reviewers = [next(reviewers_cycle) for _ in range(2)] # Assign 2 reviewers per sample
        assignment_id = uuid.uuid4()
        assignments[assignment_id] = {'sample': sample, 'reviewers': reviewers}

    return assignments

# Example usage:
assignments_ricochet = assign_reviewers(ricochet_samples)
assignments_ecole_des_loisirs = assign_reviewers(ecole_des_loisirs_samples)
```

# csv for reviewers
had minimal data: uuid (in case same sample in different sources), sample, 2 fields for the results

```python
def write_data_to_excel(assignments, reviewer):
    data = []
    for assignment_id, assignment_info in assignments:
        sample = assignment_info['sample']
        reviewers = assignment_info['reviewers']
        if reviewer in reviewers:
            data.append({'UUID': assignment_id,
                        'Echantillon': sample, 
                        'Resultat 1': '', 
                        'Resultat 2': ''})
    
    df = pd.DataFrame(data, columns=['UUID', 'Echantillon', 'Resultat 1', 'Resultat 2'])
    df.to_excel(f"{reviewer}_assignments.xlsx", index=False)   
    print("written successfully")     


write_data_to_excel(shuffled_assignments, 'Juliette')
write_data_to_excel(shuffled_assignments, 'Michel')
write_data_to_excel(shuffled_assignments, 'Karim')
```

# main csv with all the metadata of the assignements

1 row per theme, see all reviewers in same row

source of data

```python
filename = 'main_assignment_data.xlsx' # same file for both
if os.path.exists(filename):
    os.remove(filename)

def populate_data(data, assignments, source):
    for assignment_id, assignment_info in assignments.items():
        sample = assignment_info['sample']
        reviewers = assignment_info['reviewers']
        data.append({'Uuid': assignment_id,
                     'Source': source, 
                     'Echantillon': sample, 
                     'Reviewer 1': reviewers[0], 
                     "Reviewer 2": reviewers[1], 
                     'Resultat 1': '', 
                     'Resultat 2': ''})
    return data
    
def write_data_to_main_excel():
    # Define the data as a list of dictionaries

    data = []   
    data = populate_data(data, assignments_ricochet, 'ricochet')
    data = populate_data(data, assignments_ecole_des_loisirs, 'ecoles des loisirs')

    # Write the data to a CSV file
    data_frame = pd.DataFrame(data)
    data_frame.to_excel(filename, index=False)
    print("written successfully")

write_data_to_main_excel()
```
