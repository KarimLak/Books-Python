```python
import string
from collections import defaultdict
from tabulate import tabulate
import pandas as pd

def analyze_string_list(string_list, list_name):
    def get_special_character_count(string_list):
        special_character_count = {}
        for s in string_list:
            for char in s:
                if char in string.punctuation:
                    if char not in special_character_count:
                        special_character_count[char] = 1
                    else:
                        special_character_count[char] += 1
        return special_character_count

    def get_word_count(string_list):
        word_count = defaultdict(int)
        for string in string_list:
            words = string.split()
            word_count[str(len(words))] += len(words)
        return word_count

    special_character_count = get_special_character_count(string_list)
    word_count = get_word_count(string_list)

    print("################################################3")
    special_character_table = []
    for char, count in special_character_count.items():
        special_character_table.append([char, count, count / len(string_list)])

    print(tabulate(special_character_table, headers=[f"{list_name} - Character", f"{list_name} - Count", f"{list_name} - Proportion"]))

    word_count_table = []
    for word_size, count in sorted(word_count.items()):
        word_count_table.append([word_size, count, count / len(string_list)])

    print(tabulate(word_count_table, headers=[f"{list_name} - Word Size", f"{list_name} - Count", f"{list_name} - Proportion"]))


    # Create pandas dataframes from the tables
    special_character_df = pd.DataFrame(special_character_table, columns=[f"{list_name} - Character", f"{list_name} - Count", f"{list_name} - Proportion"])
    word_count_df = pd.DataFrame(word_count_table, columns=[f"{list_name} - Word Size", f"{list_name} - Count", f"{list_name} - Proportion"])

    with pd.ExcelWriter(f'{list_name}.xlsx') as writer:
        special_character_df.to_excel(writer, sheet_name=f'{list_name} - Special Characters', index=False)
        word_count_df.to_excel(writer, sheet_name=f'{list_name} - Word Count', index=False)

# Example usage
string_list = ["Hello, world!", "This is a test.", "Another example."]
analyze_string_list(ecole_des_loisir_sub_theme_list_local, "ecole_des_loisirs")
analyze_string_list(ricochet_themes_local, "ricochet")
```

```python
import spacy

# Load the English language model
nlp = spacy.load("en_core_web_md")

# Process a text
text = "I want to find synonyms for the word 'happy'"
doc = nlp(text)

# Find synonyms for each token in the text
synonyms = []
for token in doc:
    synonyms.append(token._.synonyms)

# Print the synonyms
for token, syn in zip(doc, synonyms):
    print(f"Token: {token.text}, Synonyms: {syn}")
```
